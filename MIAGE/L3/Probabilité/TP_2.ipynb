### 1.1 Minimum, Maximum et Tri
  Il est possible en python de calculer le minimum ou le maximum d'un vecteur à l'aide des commandes min et max. Par exemple:

```python
import numpy as np

X = np.random.uniform(size=10) # generate a random array from a uniform distribution
print('The array X:',X)
print('Minimum of X:', min(X))
print('Maximum of X:', max(X))
```

```python
sorted(X)
```

```python
sorted(X, reverse=True) # en ordre descendant
```

### 1.2 Histogrammes
Les histogrammes permettent de visualiser la répartition d’un ensemble de
valeurs réelles. Ils permettent de représenter par exemple un échantillons de
tirages aléatoires suivant une certaine loi. La fonction hist permet de tracer
un histogramme. Par exemple pour tracer l’histogramme d’un échantillon de
taille 1000 de loi uniforme sur [0, 1], on peut taper:

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.random.uniform(size=1000)
plt.hist(x, bins=100);
```

```python
x = np.random.uniform(size=100000)
plt.hist(x, bins=100);
```

Le deuxième argument bins donne le nombre d’intervalles de regroupement : dans
cet exemple, les valeurs de x sont regroupées dans 10 intervalles de même
taille et le diagramme affiche le nombre de valeurs à l’intérieur de chaque intervalle. Pour obtenir un histogramme plus fin, on peut entrer :

```python
plt.hist(x, bins=100);
```

On peut aussi utiliser la forme suivante (histogramme de densité), ici l’aire totale des rectangles affichés vaut 1. C’est utile si l’on veut comparer
l’histogramme avec la loi de probabilité théorique.

```python
plt.hist(x, bins=100, density=True);
```

## 2 Approximations de la loi binomiale
Dans une entreprise viticole, le remplissage des bouteilles de vin se fait à
l’aide d’une machine automatique. En régime normal chaque bouteille reçoit
exactement 75 cL de vin, mais à chaque fois la machine a une probabilité
$q = 0,03$ de se dérégler, et dans ce cas elle verse une quantité aléatoire de vin
entre 0 et 100 cL, modélisée par une loi uniforme sur $[0, 100]$. S’il y a plus de
80 cL versés, la bouteille déborde (donc le niveau de vin reste à 80 cL). On
considère que la bouteille est bien remplie si elle contient entre 74 et 76 cL
de vin.

Exercice 1. Soit X la variable aléatoire égale au niveau de vin dans une
bouteille. Quelle est la loi de X ? Écrire une fonction Bouteilles à deux
arguments $n$ et $q$ permettant de simuler un n-échantillon de cette loi. Tracer l'histogramme pour $n = 100$ et $p = 0,5$ avec l'aide de cette fonction.

```python
#n : taille de l'échantillon
#p : probabilité 

def bouteille(n,p):
    lE = [] # liste d'échantillons
    #pour n-echantllons
    for i in range(n):
        x = np.random.uniform()
        if x < 1 - p:
            lE.append(75) # 74 avec la probabilité si la machine n'est pas déréglée
        else:
            x = 100 * np.random.uniform()
            if x > 80: 
                lE.append(80) # on encadre pour pas que ça déborde
            else:
                lE.append(x)
    return lE

plt.hist(bouteille(100, 0.5), bins=100)
```

Exercice 2. Quelle est la probabilité p qu’une bouteille soit mal remplie ?
On note Y le nombre de bouteilles mal remplies dans une série de n bouteilles.
Quelle est la loi de Y ? Écrire une fonction NombreDefect à trois arguments
N, n et q qui simule N séries de n bouteilles et renvoie le nombre de bouteilles
mal remplies dans chaque série.

$P(mal / rempli) = 1 - P(74<= x <= 76)$

$P(mal / rempli) = 1 - ((1-q) + q * \frac {76 - 74}{100}) = 1 - ((1-0.003 * 0.02) = 0.0294$

Y suit la loi binômiale

```python
N = 1000
n = 100
q = 0.0294

def NombreDefect(N, n, q):
  #q: probabilite qu'une bouteille soit mal remplie
  nombre_malrempli = []
  #N series
  for i in range (N):
    nombre_malrempli.append(np.random.binomial(n,q))

  return nombre_malrempli

res = NombreDefect(N,n,q)
print(res)

plt.hist(res, range = [0, 100], bins = 100, density = True)
plt.show()
```
